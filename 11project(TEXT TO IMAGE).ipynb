```
import torch
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
from diffusers import DiffusionPipeline
```

```
pipe = DiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    safety_checker=None
)

pipe = pipe.to("cuda")
```

```
# Generate image
prompt ="A lone astronaut on an alien planet, dramatic sky, ultra-realistic"

image = pipe(
    prompt,
    num_inference_steps=30,
    guidance_scale=7.5
).images[0]
```

```
# Display image
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis("off")
plt.show()
```

### Text-to-Image Generation using Stable Diffusion

In this section, a pre-trained **Stable Diffusion v1.5** model is used to generate an image from a text prompt.  
The model converts natural language descriptions into high-quality images using a diffusion-based approach.

The following prompt was used to generate the image:

> **"A lone astronaut on an alien planet, dramatic sky, ultra-realistic"**

The model processes the text input and gradually refines random noise into a visually meaningful image that matches the description.  
This experiment demonstrates how deep learning models can be applied to creative tasks such as text-to-image generation.
